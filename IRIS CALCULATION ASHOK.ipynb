```python
from scipy import stats
F60 = df['F60']
F100= df['F100']
plt.figure(figsize=(10,8))
slope, intercept, r, p, std_err = stats.linregress(F100, F60)
def myfunc(F100):
  return slope * F100 + intercept
mymodel = list(map(myfunc, F100))
plt.scatter(F100,F60,c='r')
plt.plot(F100, mymodel, c = 'g', linewidth = 1) 
plt.legend(["bestfit line","flux"],loc="upper left")
plt.xlabel('F100 $\mu$m (MJy/sr)',fontsize=16 )
plt.ylabel('F60 $\mu$m (MJy/sr)' ,fontsize=16)
plt.tick_params(direction='in')
plt.savefig('BFI.png' )
print(slope)
print(intercept)
print(r)
print(p)
```

```python
np.random.seed(19680801)# Fixing random state for reproducibility
mu = np.mean(df['T'] )   
sigma = np.std(df['T']  )  
T = df['T'] 

plt.figure(figsize=(10,8))
n, bins, patches = plt.hist(T,  bins=20,density=True,color='springgreen')
y = ((1 / (np.sqrt(2 * np.pi) * sigma)) *
     np.exp(-0.5* (1 / sigma * (bins - mu))**2))
plt.plot(bins, y, '--',linewidth=3,c='r')
plt.legend(["Gaussian fit"],loc='upper left')
plt.xlabel(r"T$_{\mathregular{d}}$ (K)",fontsize=16)
plt.ylabel("Probability density function",fontsize=16 )
plt.savefig('GFMT.png' )
```
