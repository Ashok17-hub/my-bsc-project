```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
df = pd.read_csv('FAMAD.csv')
df
```

```python
ratio = df.F90/df.F140
df["ratio"] = ratio
df
```

```python
Temp = -57/(np.log(df.ratio*0.6**5))
df['T'] = Temp
df
```

```python
T=df['T']
c = 3.0E+8 
h = 6.62E-34
kb = 1.38E-23
lam = 140E-6
a = np.exp(h*c/(lam*kb*df['T']))
b = 1/(a-1)
planck = 2*h*c*b/lam**3
df['B(V,T)']=planck
df
```

```python
F140=df['F140']*1.25664**-19
B = df['B(V,T)']*10**16
Tau140 = F140/B
Tau140
```

```python
av = 1-np.exp(-Tau140/641.3)
df['Av']= av
df
```

```python
df.to_csv('ASHOKAKAext.csv')
```

```python
df.describe()
```

```python
su = df.F140*5.288*10**-29
D = 1682*3.0857*10**16   
Mass = 0.4*su*D**2/df['B(V,T)']
df['M']=Mass
df
```

```python
df.to_csv("ASHOKAKARImaj.csv")
```

```python
df.describe()
```

```python
from sklearn.metrics import r2_score
plt.style.use("classic")
%matplotlib inline
mu = np.mean(df['M']*10**-27 )
sigma = np.std(df['M']*10**-27 )
d = df['dma']
M = df['M']*10**-27
se = sigma/np.sqrt(28)

 
mymodel = np.poly1d(np.polyfit(d, M, 2))
mymodel1 = np.poly1d(np.polyfit(d,M, 3))
mymodel2 = np.poly1d(np.polyfit(d, M, 4))
mymodel3 = np.poly1d(np.polyfit(d, M, 5))
myline = np.linspace(0,6,100)
plt.figure(figsize=(10,8))
plt.errorbar(d, M, yerr=se,
           fmt='o',capsize=5,label='error bar',color='r' )
plt.plot(myline, mymodel(myline),ls='--',c='b',label='2nd order')
plt.plot(myline, mymodel1(myline),marker='.',markersize=7,c='k',label='3rd order')
plt.plot(myline, mymodel2(myline),c='g',label='4th order')
#lt.plot(myline, mymodel3(myline),c='black',label='5th order')
plt.legend(loc='best')
print(r2_score(M, mymodel(d)))
print(r2_score(M, mymodel1(d)))
print(r2_score(M, mymodel2(d)))
print(r2_score(M, mymodel3(d)))
print(se)
plt.xlabel("Extension (pc)",fontsize=16)
plt.ylabel(r" M$_{\mathregular{d}}$$\times 10^{27}$ (kg)",fontsize=16)
plt.savefig('AmadvsM.png' )
```
