```python
from scipy import stats
F90 = df['F90']
F140=df['F140']
slope, intercept, r, p,std_err = stats.linregress(F140,F90)
def myfunc(F140):
    return slope * F140 + intercept
mymodel = list(map(myfunc,F140))
plt.figure(figsize=(10,8))
plt.scatter(F140,F90,c='r')
plt.plot(F140, mymodel, c = 'k', linewidth = 1)

plt.legend(["bestfit line","flux"],loc="upper left")
plt.xlabel('F140 $\mu$m (MJy/sr)', fontsize = 16)
plt.ylabel('F90 $\mu$m (MJy/sr)', fontsize = 16)
plt.tick_params(direction='in') 
plt.savefig('FABF.png' )
print(slope)
print(intercept)
print(r)
print(p)
```

```python
import matplotlib.pyplot as plt  
from scipy.optimize import curve_fit
from scipy import asarray as ar,exp

x = df['d']
y = df['T']

n = len(x)                          #the number of data
mean = sum(x*y)/n                   #note this correction
sigma = sum(y*(x-mean)**2)/n        #note this correction

def gaus(x,a,x0,sigma):
    return a*exp(-(x-x0)**2/(2*sigma**2))

#popt,pcov = curve_fit(gaus,x,y)
popt,pcov = curve_fit(gaus,x,y,p0=[1,mean,sigma])

plt.plot(x,y,'b+:',label='data')
plt.plot(x,gaus(x,*popt),'ro:',label='fit')
plt.legend()
plt.title('Fig. 3 - Fit for Time Constant')
plt.xlabel('Time (s)')
plt.ylabel('Voltage (V)')
plt.show()
```
